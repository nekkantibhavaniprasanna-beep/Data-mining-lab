# Experiment 5: Classification using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

X = data.iloc[:, :-1]
y = data.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```

# Experiment 7: Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

dataset = [
    ['milk', 'bread'],
    ['bread', 'butter'],
    ['milk', 'bread', 'butter']
]

te = TransactionEncoder()
te_data = te.fit(dataset).transform(dataset)

df = pd.DataFrame(te_data, columns=te.columns_)

frequent = apriori(df, min_support=0.5, use_colnames=True)
print(frequent)
```

# Experiment 8: FP Growth


```python
from mlxtend.frequent_patterns import fpgrowth

fp = fpgrowth(df, min_support=0.5, use_colnames=True)
print(fp)
```

# Experiment 9: K-Means Clustering

```python
from sklearn.cluster import KMeans
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

kmeans = KMeans(n_clusters=3)
kmeans.fit(data.iloc[:, :-1])

print(kmeans.labels_[:10])
```

# Experiment 10: Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

Z = linkage(data.iloc[:, :-1], method='ward')
print(Z[:5])
```
